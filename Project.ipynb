<br><br>
یادگیری گروهی یا جمعی $(Ensemble~Machine~Learning)$، با ترکیب پیش‌بینی‌ها و دنبال‌کردن دانش گروه، یک دسته‌بندی نهایی یا خروجی با پیش‌بینی بهتر را ارائه می‌دهد. روش های یادگیری گروهی را می‌توان به دسته‌های مختلفی دسته‌بندی کرد از جمله ترتیبی $(Sequential)$، موازی $(Parallel)$، همگن $(Homogenous)$ و ناهمگن $(Heterogeneous)$. روش <b>تقویت گرادیان</b> $(Gradient~Boosting)$ از نوع ترتیبی و روش <b>تقویت گرادیان شدید</b> $(Extreme~gradient~ boosting)$ از نوع  موازی است که در ادامه به آن ها اشاره می کنیم. 
<br><br>

<b>1-1) تقویت گرادیان</b>
<br><br>
تقویت گرادیان توسط جروم اچ فریدمن $(Jerome~H.~Friedman)$ با تکیه بر کار لئو برایمن $(Leo~Breiman)$ توسعه داده شد و به  دسته‌ای از الگوریتم‌های یادگیری ماشینی گروهی  اشاره دارد که می‌توانند برای مسائل <b>دسته‌بندی</b> یا <b>رگرسیون</b> استفاده شوند. اگر مدل تقویت گرادیان مبتنی بر درخت تصمیم باشد به آن ماشین تقویت گرادیان یا به اختصار $GBM$ نیز می گویند.
<br><br>
    مدل تقویت گرادیان <b>ترکیبی خطی</b> از یک سری <b>مدل‌های ضعیف</b> است که به صورت دنباله وار برای ایجاد یک <b>مدل نهایی قوی</b> ساخته شده‌است و در هر مرحله خطاهای پیش بینی کننده قبلی خود را تصحیح می کنند و به جای تغییر وزن نقاط داده مانند $AdaBoost$، یادگیری مدل ها براساس خطاهای باقیمانده پیش بینی کننده قبلی است. عملکرد آن همواره از الگوریتم‌های اساسی یا ضعیف (مثلا درخت تصمیم) یا روش‌هایی مانند جنگل تصادفی بهتر است، اما این عملکرد تا حدی از مشخصات داده‌های ورودی تأثیر می‌پذیرد.
<br><br>
    این الگوریتم از دو مفهوم مهم <b>گرادیان کاهشی</b> $(Gradient~Descent)$ و <b>تقویت</b> $(Boosting)$ استفاده می کند که در ادامه به آن اشاره می کنیم.
<br><br>
$$Gradient~Boosting = Gradient~Descent + Boosting$$

<b>3 & 2) بیان ریاضیات روش پایه و روش جدید</b>
<br><br>
ابتدا به بیان ریاضی تقویت می پردازیم. همانطور که گفتیم تقویت در یادگیری ماشین متعلق به خانواده مدل های گروهی است که در آن تمرکز اصلی بر کاهش خطا است. یعنی باید ابتدا یک مدل ساخت، باقیمانده آن را پیدا کرد و روی باقیمانده مدل دیگری ساخت. این فرآیند هر چند بار که لازم است تکرار می شود. از نظر ریاضی می توان آن را به صورت زیر نشان داد.
فرض کنید $f_i(x)$ مدل ما می باشد، $y$ مقدار واقعی، $\hat{\gamma}$ مقدار پیش بینی شده و $L$ تابع زیان است. اولین مدل $f_0 (x)$ به صورت زیر در $\{x_i , y_i \}$ ساخته شده می شود:
<br><br>
$$f_i (x) = argmin_{\hat{\gamma}}L(y,\hat{\gamma})$$
<br>
اکنون باقیمانده ( $\theta=(y-\hat{\gamma})$ ) را محاسبه کرده و مدل دوم $h_1 (x)$ را روی $\{x_i , \theta_i \}$  می سازیم. $h_1 (x)$ را به $f_0 (x)$ اضافه کرده و مدل جدید بهبود یافته $h_1 (x)$ ساخته می شود.
<br><br>
$$f_1 (x)=f_0 (x) + h_1 (x)$$
<br>
با تکرار پی در پی روند بالا، شکل تعمیم یافته معادله فوق در تکرار $m$ به صورت زیر خواهد بود:
<br><br>
$$f_m (x)=f_{m-1} (x) + h_m (x)$$
<br>
معادله فوق، رابطه عمومی برای الگوریتم تقویت است. راه های مختلفی برای تصمیم گیری درباره این که چه نسبتی از $h_m (x)$ به $f_{m-1} (x)$ اضافه شود وجود دارد که با این فرض معادله فوق به شکل زیر خواهد بود.
<br><br>
$$f_m (x)=f_{m-1} (x) + \alpha h_m (x) \hspace{1cm}(*)$$
<br>
روش های مختلفی برای محاسبه $\alpha$ وجود دارد مانند میانگین، میانگین وزنی، $adaboost$ یا استفاده از تقویت گرادیان، که در ادامه تقویت گرادیان را مورد بررسی قرار می دهیم.
اصطلاح گرادیان در تقویت گرادیان از مفهوم کاهش گرادیان ناشی می شود. روش مبتنی بر کاهش گرادیان برای تعیین اندازه آلفا یا گام استفاده می شود. برای محاسبه آلفا، در تکرار $m$، ابتدا شبه باقیمانده ( $\theta_{i_m}$ ) محاسبه می شود و مدل جدید $h_m (x)$ بر روی $\{x_i , \theta_{i_m} \}$ ساخته می شود. شبه باقیمانده به صورت زیر محاسبه می شود:
<br><br>
$$\theta_{i_m} = \frac{\partial{L}}{\partial{f(x_i)}}(y_i , f(x_i))|_{f(x)=f_{m-1}(x)}$$
<br>
حالا $\alpha$ را طوری محاسبه می کنیم که تابع زیان کمینه شود.
<br><br>
$$\alpha = argmin_{\alpha}L(y_i , f_{m-1}(x_i)+\alpha h_{m}(x))$$
<br>
حالا مقادیر $\alpha$ و $h_m (x)$ را در رابطه ( $*$ ) قرار می دهیم تا $f_m (x)$ بدست آید.
اگر مدل مبتنی بر درخت تصمیم باشد به آن ماشین تقویت گرادیان یا به اختصار $GBM$ می گویند، یعنی اگر $f(x)$ و $h(x)$ درختان $CART$ باشند آنگاه برای درختی با  $T$ برگ، مدل $h_m (x)$ را می توان به صورت زیر نوشت:
<br><br>
$$h_m (x)=\sum_{j=1}^{T}b_{j_m}l_{R_{j_m}}(x)$$
<br>
$b_{j_m}$ مقدار پیش بینی شده (میانگین، میانه، حداکثر رای و ...) در ناحیه $R_{j_m}$ (برگ $j$) است. اگر $h_m (x)$ برای یک درخت در معادله تقویت گرادیان قرار داده شود، $\alpha$ و $b_{j_m}$ وجود خواهند داشت. در $GBM$، $\alpha$ و $b_{j_m}$ با هم ترکیب می شوند تا برای هر برگ نرخ گام به دست آید. بنابراین، $T$ آلفا (نرخ گام) در درختی با $T$ برگ وجود خواهد داشت. معادلات $GBM$ به صورت زیر هستند:
<br><br>
$$f_m (x)=f_{m-1} (x) + \sum_{j=1}^{T}\alpha_{j_m}l_{R_{j_m}}(x)$$
<br>
$$\alpha_{j_m}=argmin_{\alpha}\sum_{j\in R_{j_m}}L(y_i , f_{m-1}(x_i)+\alpha)$$
<br>
همانطور که دیدیم محاسبه نرخ گام در $GBM$ به دو مرحله نیاز دارد 1) محاسبه شبه باقیمانده 2) محاسبه $\alpha$. همچنین، احتمال بیش برازش در $GBM$ وجود دارد که با افزودن ترم منظم سازی کاهش می یابد.
الگوریتم تقویت گرادیان شدید $(XGBoost)$ دو مرحله را برای محاسبه نرخ گام در یک مرحله ترکیب می‌کند و برای جلوگیری از بیش براز، ترم منظم سازی را در تابع زیان اضافه می‌کند. در تقویت گرادیان شدید، درخت در هر تکرار به گونه‌ای ساخته می‌شود که بهره $(gain)$ در هر تقسیم حداکثر باشد. 
همانطور که بیان شد تابع زیان در الگوریتم تقویت گرادیان یک ترم منظم سازی $omega$ دارد. این تابع زیان برای محاسبه حداکثر بهره استفاده می شود که می تواند به طور مستقیم در فرآیند درخت سازی در حین تقسیم هر گره مورد استفاده قرار گیرد. بنابراین، اینگونه است که فرآیند دو مرحله ای در $GBM$ با تنظیم به یک مرحله کاهش می یابد.
حال به بررسی نحوه محاسبه بهره در ساختمان درخت تقویت گرادیان شدید می پردازیم. از تابع زیان که باید به کمینه شود شروع می کنیم.
<br><br>
$$Loss=\sum_{i=1}^{n}L(y_i , f_{m-1}(x_i)+h_{m}(x_i))+\sum_{j=1}^{T}\Omega(h_{m}(x_i))$$
<br>
می‌تواند چندین نوع تابع زیان ( $L$ در معادله فوق ) مانند $MSE$ برای رگرسیون یا آنتروپی متقاطع برای دسته بندی وجود داشته باشد. برای تشکیل یک معادله کلی بدون توجه به تابع زیان، با نوشتن بسط مرتبه دو سری تیلور برای تابع $L$ داریم:
<br><br>
$$Loss=\sum_{i=1}^{n}[ L(y_i , f_{m-1}(x_i))+\frac{\partial{L}}{\partial{f_{m-1}(x_i)}}(y_i , f_{m-1}(x_i))h_{m}(x_i)]$$
<br>
$$+\frac{1}{2}\frac{\partial^2{L}}{\partial^2{f_{m-1}(x_i)}}(y_i , f_{m-1}(x_i))h_{m}^{2}(x_i)+\sum_{j=1}^{T}\Omega(h_{m}(x_i))$$
<br>
که عبارت ثابت تولید شده یعنی $L(y_i , f_{m-1}(x_i))$ حذف می شود زیرا در محاسبه بهره در حین تقسیم یک گره در درخت نقشی ندارد. با قرار دادن $g_i =\frac{\partial{L}}{\partial{f_{m-1}(x_i)}}(y_i , f_{m-1}(x_i))$ و $k_i =\frac{\partial^2{L}}{\partial^2{f_{m-1}(x_i)}}(y_i , f_{m-1}(x_i))h_{m}^{2}(x_i)$ و $\Omega(h_{m_T}) =\sum_{j=1}^{T}\Omega(h_{m}(x_i))$  داریم:
<br><br>
$$Loss= \sum_{i=1}^{n}[g_i h_m(x_i)+\frac{1}{2}k_i h_{m}^{2}(x_i)]+\Omega(h_{m_T})$$
<br>
ترم منظم سازی امگا را می توان به صورت زیر نوشت:
<br><br>
$$\Omega(h_{m_T})=\epsilon T + \frac{1}{2}\lambda \sum_{j=1}^{T}b_{j_m}^{2}$$
<br>
با جایگزینی مقدار $h_m (x)$ به عنوان معادله درخت و حذف $m$ (تعداد تکرار) برای مرتب بودن، داریم:
<br><br>
$$Loss= \sum_{i=1}^{n}[g_i \sum_{j=1}^{T}b_{j_m}l_{R_j}(x_i)+\frac{1}{2}k_i (\sum_{j=1}^{T}b_{j_m}l_{R_j}(x_i))^2 ]+\epsilon T + \frac{1}{2}\lambda \sum_{j=1}^{T}b_{j_m}^{2}$$
<br>
در نتیجه:
<br><br>
$$Loss= \sum_{j=1}^{T}[(\sum_{i\in I_j}g_i)b_j +\frac{1}{2}(\sum_{i\in I_j}(R_j +\lambda))b_{j}^{2}]+\epsilon T$$
<br>
با ساده سازی معادله قبل داریم:
<br><br>
$$Loss= \sum_{j=1}^{T}[G_j b_j + \frac{1}{2}(H_j + \lambda) b_{j}^{2}]+\epsilon T$$
<br>
هدف ما به حداقل رساندن این تابع زیان است. همانطور که از این معادله مشخص است عبارت داخل سیگما معادله یک سهمی به صورت $ax+bx^2$ است و کمترین مقدار سهمی در $x=-\frac{a}{2b}$ رخ خواهد داد. در مورد معادله قبل کمترین مقدار در $b_j = -\frac{G_j}{H_j+\lambda}$ رخ می دهد. با جایگزینی $b_j$ در تابع زیان، به تابع زیان بهینه تبدیل می شود:
<br><br>
$$Loss= \sum_{j=1}^{T}[-\frac{G_j^2}{H_j+\lambda} + \frac{1}{2}(H_j + \lambda) (\frac{G_j}{H_j+\lambda})^{2}]+\epsilon T$$
<br>
$$Loss= \sum_{j=1}^{T}[-\frac{G_j^2}{H_j+\lambda} + \frac{1}{2}\frac{G_j^2}{H_j+\lambda}]+\epsilon T$$
<br>
$$Loss= -\frac{1}{2}\sum_{j=1}^{T}[\frac{G_j^2}{H_j+\lambda}]+\epsilon T$$
<br>
هنگامی که یک برگ به دو برگ (چپ $L$ و $R$ راست) تقسیم می شود، بهره به صورت زیر در می آید:
<br><br>
$$Gain = Loss~~before~~split - Loss~~after~~split$$
<br>
$$=\epsilon - \frac{1}{2}\frac{(G_L + G_R )^2}{H_L + H_R + \lambda}-[\frac{-G_L^2}{2(H_L + \lambda)}+\epsilon-\frac{G_R^2}{2(H_R + \lambda)}+\epsilon]$$
<br>
$$=-\epsilon + \frac{1}{2}[\frac{G_L^2}{H_L + \lambda}+\frac{G_R^2}{H_R + \lambda}-\frac{(G_L + G_R )^2}{H_L + H_R + \lambda}]$$
<br>
با استفاده از این بهره، الگوریتم تقویت گرادیان شدید، ساختار درختی را می سازد.

<br><br>
پارامتر loss (تابع زیان): مقدار تابع زیان یکی از مقادیر $exponential$ ،$deviance$ و $log$_$loss$ (حالت پیش فرض) می تواند باشد.

```python

def predict_diabetes(Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DPF, Age):
    preg = int(Pregnancies)
    glucose = float(Glucose)
    bp = float(BloodPressure)
    st = float(SkinThickness)
    insulin = float(Insulin)
    bmi = float(BMI)
    dpf = float(DPF)
    age = int(Age)

    x = [[preg, glucose, bp, st, insulin, bmi, dpf, age]]
    x = scaler.transform(x)

    return gb.predict(x)
```
